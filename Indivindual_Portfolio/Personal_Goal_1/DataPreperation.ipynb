Some values in columns likes Beach have a whitespace either at the begging of the word either at the end. The result is that after the merge it is assigned as NaN due to the whitespace so sometimes it is a good habit to strip the whitespaces to make a merge successfull. Take for example the Porto palo beach.

```python
#It doesn't return anything
beaches_wwf_materials[beaches_wwf_materials["Beach"] == "Porto palo"]
```

```python
#perform the strip
beaches_wwf_materials["Beach"] = beaches_wwf_materials["Beach"].str.strip()
```

```python
#now it returns
beaches_wwf_materials[beaches_wwf_materials["Beach"] == "Porto palo"]
```

```python
#merge the datasets according to the beach name based on relationship
merged_df= pd.merge(
    reviews_dataset,
    beaches_wwf_materials,
    left_on = "Original_Beach_Name", # Use 'Original_Beach_Name' from reviews_dataset
    right_on = "Beach",         # Use 'Spiaggia' from beaches_wwf_materials
    how = "left"
)
```

```python
#chech if the merged happend successfully
merged_df.head(5)
```

```python
merged_df.info()
```

```python
#drop the unnecessary columns
df_column_cleaned = merged_df.drop(columns=["Beach","Search_lat","Search_lng",
                                    "Sesso animale", "Animale impigliato nei rifiuti (Descrizione)", "Durante il tracciamento - registrazione, ti allontani dal punto preimpostato di 100 m (Descrizione)",
                                    "Descrizione degli animali", "Età dell'animale", "Animale impigliato nei rifiuti",
                                    "Vivo o morto", "Eventi o situazioni che potrebbero aver influenzato la registrazione",
                                    "Eventi o situazioni che potrebbero aver portato all'aumento dei rifiuti sulla spiaggia", "Fenomeno meteorologico"])
```

```python
# As we can see there is inconsistency regarding the missing values but we need
#to save as much information is possible
df_column_cleaned.info()
```

```python
# a copy is created
final_df = df_column_cleaned.copy()
```

```python
# removing the rows that they missing rating values
final_df = final_df.dropna(subset=["rating"])
```

```python
#check again the info
final_df.info()
```

```python
#check which columns have empty values
final_df[final_df["Trash amount"].isna()]
```

```python
# As it can be seen from the info method above the Trash Amount is Object which it should be numeric
# So we convert into numeric
final_df["Trash amount"] = pd.to_numeric(final_df["Trash amount"], errors= "coerce")
```

```python
# Drop columns that have NaN values regarding the Trash amount
final_df = final_df.dropna(subset=["Trash amount"])
```

```python
# check it again
final_df.info()
```

```python
final_df["Period"].unique()
```

```python
#Bring the names into the same consistency
final_df["Period"] = final_df["Period"].replace({
"Winter 2024-2025" : "Winter 2024",
"Winter 2025-2026" : "Winter 2025"  })

# Strip column names
final_df.columns = final_df.columns.str.strip()
```

```python
# as it can be seen the data type of the length column is object but it should be number
# so convert it into numeric
final_df["length"] = (
    final_df["length"]
    .str.replace(",", ".", regex=False)
    .astype(float)
)
```

```python
# Saving merged dataset as .csv
beaches_wwf_materials.head(4)
final_df.head()
final_df.to_csv("merged_dataset_v2.csv", index=True)
```
